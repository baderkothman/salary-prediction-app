# load dataset

```python
import pandas as pd
import numpy as np

raw_path ="..\\data\\raw\\ds_salaries.csv"

df = pd.read_csv(raw_path)

df.head()
```

# check basic dataset info

```python
df.shape
```

```python
df.info
```

```python
df.describe(include='all')
```

# Standardize column names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df.columns
```

# check for missing values

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values
```

```python
missing_percent = (df.isnull().mean() * 100).sort_values(ascending=False)
missing_percent
```

```python
df = df.dropna()
df.shape
```

# check duplicates

```python
df.duplicated().sum()
df = df.drop_duplicates()

df.shape
```

# clean text columns

```python
categorical_columns = [
    "experience_level",
    "employment_type",
    "job_title",
    "employee_residence",
    "company_location",
    "company_size",
]

for col in categorical_columns:
    df[col] = df[col].astype(str).str.strip()

df.head()
```

# check for unique values

```python
for col in categorical_columns:
    print(f"\n{col}")
    print(df[col].value_counts())
```

# rename coded values

```python
experience_map = {
    "EN": "Entry-level",
    "MI": "Mid-level",
    "SE": "Senior-level",
    "EX": "Executive-level",
}

employment_map = {
    "FT": "Full-time",
    "PT": "Part-time",
    "CT": "Contract",
    "FL": "Freelance",
}

company_size_map = {"S": "Small", "M": "Medium", "L": "Large"}

df["experience_level"] = df["experience_level"].replace(experience_map)
df["employment_type"] = df["employment_type"].replace(employment_map)
df["company_size"] = df["company_size"].replace(company_size_map)

df.head()
```

# clean numeric columns

```python
numeric_columns = ["work_year", "salary", "salary_in_usd", "remote_ratio"]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df = df.dropna(subset=numeric_columns)

df[numeric_columns].info()
```

# remove invalid salary rows

```python
df = df[df["salary_in_usd"] > 0]
df = df[df["salary"] > 0]

df.shape
```

# check remote ratio values

```python
df["remote_ratio"].value_counts().sort_index()
```

```python
remote_map = {0: "On-site", 50: "Hybrid", 100: "Remote"}

df["work_setting"] = df["remote_ratio"].map(remote_map)

df[["remote_ratio", "work_setting"]].head()
```

# decide ml columns

```python
model_df = df[
    [
        "work_year",
        "experience_level",
        "employment_type",
        "job_title",
        "employee_residence",
        "remote_ratio",
        "work_setting",
        "company_location",
        "company_size",
        "salary_in_usd",
    ]
].copy()

model_df.head()
```

```python
model_df.info()
```

```python
model_df.isnull().sum()
```

```python
model_df.duplicated().sum()
```

```python
model_df = model_df.drop_duplicates()
model_df.shape
```

# save new processed data

```python
processed_path = "../data/processed/cleaned_salaries.csv"

model_df.to_csv(processed_path, index=False)

print(f"Cleaned dataset saved to: {processed_path}")
```
